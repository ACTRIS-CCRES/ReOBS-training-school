# Hands on ReOBS

## Objective
The purpose of this notebook is to explore the **ReOBS** dataset, produced as part of the National Facility process and initially accepted for **labelling step 1b**.  
These files were generated within the framework of data quality assessment (labelling step 1b), focusing on performance evaluation over a two-year period.

## Exploring a ReOBS file with xarray
This notebook illustrates a first exploration of a NetCDF ReOBS file (or similar).
We will:
- Open the file with **xarray**
- Inspect the main characteristics: dimensions, coordinates, available variables, global attributes
- Check the time coverage (start/end dates) and **time_bnds** to illustrate temporal resolution
- Plot the vertical grid
- Make a simple plot of a few variables

### ReOBS tool
To assess data quality, CCRES makes the use of the **ReOBS tool** (https://reobs.aeris-data.fr/en/welcome/, Chiriaco et al., 2018), which synthesizes all Cloudnet geophysical and instrument products into a single `.nc` file with a 1-hour temporal resolution and a common vertical grid.  
Additional quality controls (QCs) are applied, generally using more restrictive flags/status indicators from the original products.  
For further details on ReOBS methodology and the QC procedures, please refer to the **ReOBS ATBD**: https://drive.google.com/file/d/1EyAEPVkyErT9fDAAf0Uy3h_rjA_aH-eU/view?usp=sharing.

```python
# Import packages needed for the present notebook
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import xarray as xr
from matplotlib.colors import LogNorm

# Define data path
data_dir = Path("/data/ACTRIS-CCRES/REOBS")
```

```python
station = "lindenberg" # chose your station
reobs_file = list(data_dir.glob(f"{station}_reobs_ccres_2D_*.nc"))[0]
if reobs_file:
    ds = xr.open_dataset(reobs_file)
else:
    print(f"No ReOBS file found for station {station}")
```

#### 3 - Simple plots of selected variables

```python
date_start = "2025-07-01 00:00"
date_end = "2025-07-31 23:59"
ds2plot = ds.sel(time=slice(date_start, date_end))
```

##### Plot air temperature and relative humidity time series

```python
fig, ax = plt.subplots(figsize=(14,7))

ax.plot(ds2plot.time, ds2plot.ta_ws, lw=2, color="r", label=ds2plot.ta_ws.long_name,)
ax.set_ylabel(f"[{ds2plot.ta_ws.units}]")
ax.legend(loc="upper left")
ax.set_title(ds2plot.station_name.long_name)

ax02 = ax.twinx()
ax02.plot(ds2plot.time, ds2plot.rh_ws, lw=2, color="b", label=ds2plot.rh_ws.long_name)
ax02.set_ylabel(f"[{ds2plot.rh_ws.units}]")
ax02.legend(loc="upper right")
ax.grid(ls="--")

plt.show()
```

##### Plot rain info time series

```python
def derive_rainfall_info(da: xr.DataArray) -> set[np.ndarray, np.ndarray]:
    """
    Get rainfall info from rain measurements.

    Parameters
    ----------
    da : xr.DataArray
        Rainfall DataArray.

    Returns
    -------
    np.ndarray
        Hourly rain.
    np.ndarray
        Total amount of rain.
        
    """
    hourly_rain = np.zeros_like(da, dtype=float)
    # Calculation of hourly rainfall with correction for resets at midnight
    for i in range(1, da.size):
        if da.time.dt.hour[i] == 0:  # if midnight
            hourly_rain[i] = da.values[i]  # just take the value measured at 00h
        else:
            hourly_rain[i] = max(
                0, da.values[i] - da.values[i - 1]
            )  # standard difference
    # Calculation of the total amount, taking account of corrections
    total_amount = np.nancumsum(hourly_rain)
    
    return hourly_rain, total_amount
```

```python
# PLOT RAIN
hourly_rain, total_amount = derive_rainfall_info(ds2plot.ra_ws_max)

fig, ax = plt.subplots(figsize=(14,7))

ax.plot(ds2plot.time, hourly_rain, lw=2, color="k", label="hourly rainfall")
ax.set_ylabel("Rainfall rate [mm/h]")
ax.legend(loc="upper left")
ax.set_title(ds2plot.station_name.long_name)

ax02 = ax.twinx()
ax02.plot(ds2plot.time, total_amount, lw=2, color="DarkOrange", label="cumulative hourly rainfall")
ax02.set_ylabel("Rainfall acc. [mm]")
ax02.legend(loc="upper right")
ax.grid(ls="--")

plt.show()
```

##### Plot reflectivity, mwr temperature profile, ALC bck (2D plots)

- Chose : reflectivity, mwr_temperature or  bck_alc

```python
var = "reflectivity" 
decode_var = {
    "reflectivity": {
        "var_name": "zh_dcr",
        "vmin": -50, 
        "vmax": 30,
    },
    "mwr_temperature":{
        "var_name": "ta_mwr",
        "vmin": 200, 
        "vmax": 300,
    },
    "bck_alc": {
        "var_name": "bck_smooth_alc",
        "vmin": 1e-7,
        "vmax": 1e-4
    },
}

fig, ax = plt.subplots(figsize=(14,7))

if var == "bck_alc":
    im = ax.pcolormesh(
        ds2plot.time, 
        ds2plot.altitude/1000, 
        ds2plot[decode_var[var]["var_name"]].T, 
        cmap=plt.get_cmap("jet"), 
        norm=LogNorm(vmin=decode_var[var]["vmin"], vmax=decode_var[var]["vmax"]),
    )
else:
    im = ax.pcolormesh(
        ds2plot.time, 
        ds2plot.altitude/1000,
        ds2plot[decode_var[var]["var_name"]].T, cmap=plt.get_cmap("jet"), 
        vmin=decode_var[var]["vmin"], 
        vmax=decode_var[var]["vmax"],
    )
    
cbar = plt.colorbar(im)
cbar.ax.set_ylabel(f"{ds2plot[decode_var[var]["var_name"]].long_name} [{ds2plot[decode_var[var]["var_name"]].units}]")

ax.set_title(ds2plot.station_name.long_name)
ax.set_ylabel("Altitude [km]")
ax.set_ylim(0,12.5)
ax.grid(ls="--")

plt.show()
```
